This script will map network changes from the original Contra Costa network to the adjusted network, for use in the poster.

```python
import os
import pandas as pd
import numpy as np
from pyomo.environ import *
import pyomo.opt as pyopt

import geopandas as gpd
import pandas as pd
import numpy as np
from statistics import mean

import itertools
import osmnx as ox
import folium
import networkx as nx
import plotly as plotly
```

```python
# load the original and adjusted graphs
adjusted = ox.load_graphml(os.path.join(os.getcwd(), 'data', 'graphs', 'graph_walk_cfcontracosta.graphml'))
original = ox.load_graphml(os.path.join(os.getcwd(), 'data', 'graphs', 'graph_walk_contracosta.graphml'))
```

```python
#pt = (37.9625591867685, -122.3453201572392)
#pt = (37.9327196688036, -122.32681823855924)
pt = (37.93238911989639, -122.34786416748183)
bbox = ox.utils_geo.bbox_from_point(pt, dist = 500)
cf = ('''["highway"]["area"!~"yes"]["access"!~"private"]
                ["highway"~"pedestrian|living_street|tertiary|secondary|primary|residential"]
                ["service"!~"private"]{}''').format(ox.settings.default_access)
cf_walk = ('''["highway"]["area"!~"yes"]
["access"!~"private"]["highway"!~"abandoned|bus_guideway|construction
|cycleway|motor|planned|platform|proposed|raceway"]
["foot"!~"no"]["service"!~"private"]{}''').format(ox.settings.default_access)
cf_both = ('''["highway"]["area"!~"yes"]
["access"!~"private"]["highway"!~"abandoned|bus_guideway|construction
|cycleway|motor|planned|platform|proposed|raceway"]
["foot"!~"no"]["service"!~"private"]["highway"]["area"!~"yes"]["access"!~"private"]
                ["highway"~"pedestrian|living_street|tertiary|secondary|primary|residential"]
                ["service"!~"private"]{}''').format(ox.settings.default_access)
```

```python
adjusted_bbox = ox.graph.graph_from_bbox(bbox[0], bbox[1], bbox[2], bbox[3], custom_filter = cf, simplify = True)
original_bbox = ox.graph.graph_from_bbox(bbox[0], bbox[1], bbox[2], bbox[3], custom_filter = cf_walk, simplify = True)
both_bbox = ox.graph.graph_from_bbox(bbox[0], bbox[1], bbox[2], bbox[3], custom_filter = cf_both, simplify = True)
```

```python

m = ox.folium.plot_graph_folium(adjusted_bbox, 
                               graph_map=None, 
                               popup_attribute=None, 
                               tiles='cartodbpositron', color = 'lightgreen')
m = ox.folium.plot_graph_folium(original_bbox, 
                               graph_map=m, 
                               popup_attribute=None, 
                               tiles='cartodbpositron',
                           color = 'red')
m = ox.folium.plot_graph_folium(both_bbox, 
                               graph_map=m, 
                               popup_attribute=None, 
                               tiles='cartodbpositron', color = 'black')
m
```
