This script will map walking graph and additional public transit graph, for use in the poster.

```python
import os
import pandas as pd
import numpy as np
from pyomo.environ import *
import pyomo.opt as pyopt

import geopandas as gpd
import pandas as pd
import numpy as np
from statistics import mean

import itertools
import osmnx as ox
import folium
import networkx as nx
import plotly as plotly
```

```python
# load the walking and public transit graphs
walk = ox.load_graphml(os.path.join(os.getcwd(), 'data', 'graphs', 'graph_walk_cfcontracosta.graphml'))
transit = ox.load_graphml(os.path.join(os.getcwd(), 'data', 'graphs', 'graph_walk_cf_transit_contracosta.graphml'))
```

```python
pt = (37.9465492516322, -122.33056734376855)
bbox = ox.utils_geo.bbox_from_point(pt, dist = 400)
cf = ('''["highway"]["area"!~"yes"]["access"!~"private"]
                ["highway"~"pedestrian|living_street|tertiary|secondary|primary|residential"]
                ["service"!~"private"]{}''').format(ox.settings.default_access)
```

```python
walk_bbox = ox.graph.graph_from_bbox(bbox[0], bbox[1], bbox[2], bbox[3], custom_filter = cf, simplify = True)
```

```python
transit_bbox = ox.truncate.truncate_graph_bbox(transit, bbox[0], bbox[1], bbox[2], bbox[3])
```

```python
# Note that connecting edges will need to be colored by hand

m = ox.folium.plot_graph_folium(transit_bbox, 
                               graph_map=None, 
                               popup_attribute=None, 
                               tiles='cartodbpositron', color = 'blue')
m = ox.folium.plot_graph_folium(walk_bbox, 
                               graph_map=m, 
                               popup_attribute=None, 
                               tiles='cartodbpositron',
                           color = 'red')
m
```
